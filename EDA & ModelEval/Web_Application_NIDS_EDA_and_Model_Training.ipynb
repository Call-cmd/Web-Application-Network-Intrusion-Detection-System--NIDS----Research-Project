# Train/Test Pipeline w/ Weight balancing/SMOTE

```python
# ml_models_comparison.py

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import lightgbm as lgb
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# Helper functions
# ---------------------------
def print_eval(y_true, y_pred, labels=None):
    print("Classification report:\n")
    print(classification_report(y_true, y_pred, digits=4, labels=labels))
    print("Macro F1: {:.4f}".format(f1_score(y_true, y_pred, average='macro')))
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    return cm

def plot_confusion(cm, classes, title="Confusion matrix"):
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=classes, yticklabels=classes, cmap='Blues')
    plt.ylabel('True')
    plt.xlabel('Predicted')
    plt.title(title)
    plt.show()

# ---------------------------
# Load your data here
# ---------------------------

X = mod_df.drop(['Label'], axis=1)
y = mod_df['Label']


# Train/test split
# ---------------------------

test_size = 0.2
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, stratify=y, random_state=random_state)


# Compute sample weights from class frequencies (useful for XGBoost/LightGBM)
# ---------------------------
unique, counts = np.unique(y_train, return_counts=True)
class_freq = dict(zip(unique, counts))
print("Train class frequencies:", class_freq)
n_total = len(y_train)
class_weights = {cls: n_total/(len(unique)*cnt) for cls, cnt in class_freq.items()}  # inverse frequency
# sample_weight array for training examples
sample_weight = np.array([class_weights[int(lbl)] for lbl in y_train])


# 1) XGBoost (multiclass)
# ---------------------------
xgb_model = XGBClassifier(
    objective='multi:softprob',
    num_class=len(unique),
    eval_metric='mlogloss',
    use_label_encoder=False,
    n_estimators=16,
    max_depth=6,
    learning_rate=0.1,
    random_state=random_state,
    n_jobs=-1
)

print("\nTraining XGBoost with sample weights (handles imbalance via sample_weight)...")
xgb_model.fit(X_train, y_train, sample_weight=sample_weight, eval_set=[(X_test, y_test)], verbose=False)

y_pred_xgb = xgb_model.predict(X_test)
cm_xgb = print_eval(y_test, y_pred_xgb, labels=unique)
plot_confusion(cm_xgb, classes=[str(c) for c in unique], title="XGBoost Confusion Matrix")

# 2) LightGBM (multiclass)
# ---------------------------
# You can pass class_weight parameter mapping or 'balanced' for lightgbm.Dataset,
# but in sklearn API, use class_weight param in LGBMClassifier or sample_weight.
lgb_model = lgb.LGBMClassifier(
    n_estimators=16,
    learning_rate=0.1,
    max_depth=-1,
    random_state=42,
    verbosity=-1   # suppresses training logs
)

# Train with sample weights
print("\nTraining LightGBM with sample weights...")
lgb_model.fit(
    X_train, y_train,
    sample_weight=sample_weight,
    eval_set=[(X_test, y_test)],
    callbacks=[]  # you could also add callbacks like early stopping here
)

y_pred_lgb = lgb_model.predict(X_test)
cm_lgb = print_eval(y_test, y_pred_lgb, labels=unique)
plot_confusion(cm_lgb, classes=[str(c) for c in unique], title="LightGBM Confusion Matrix")


# 3) KNN (distance-weighted)
# ---------------------------
# KNN needs scaling and benefits from oversampling to address imbalance
knn_pipeline = ImbPipeline([
    ('scaler', StandardScaler()),
    #('smote', SMOTE(random_state=random_state, n_jobs=-1)),   # optionally enable SMOTE for KNN
    ('knn', KNeighborsClassifier(n_neighbors=7, weights='distance', n_jobs=-1))
])

print("\nTraining KNN (with scaling). If dataset large, this will be slow at predict time.")
knn_pipeline.fit(X_train, y_train)
y_pred_knn = knn_pipeline.predict(X_test)
cm_knn = print_eval(y_test, y_pred_knn, labels=unique)
plot_confusion(cm_knn, classes=[str(c) for c in unique], title="KNN Confusion Matrix")


# 4) Kernel SVM (RBF)
# ---------------------------
# SVM needs scaling. Use 'class_weight' to handle imbalance OR SMOTE on training set.
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='rbf', C=1.0, gamma='scale', class_weight='balanced', probability=False))
])

print("\nTraining Kernel SVM (RBF) with class_weight='balanced'. Note: SVM can be slow on big datasets.")
svm_pipeline.fit(X_train, y_train)
y_pred_svm = svm_pipeline.predict(X_test)
cm_svm = print_eval(y_test, y_pred_svm, labels=unique)
plot_confusion(cm_svm, classes=[str(c) for c in unique], title="Kernel SVM Confusion Matrix")

# cross-validation comparison (macro F1)
# ---------------------------
print("\nCross-validation macro-F1 (StratifiedKFold) for each model (use smaller folds if dataset big)...")

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

def cv_score_model(model, X, y, folds=3, random_state=42):
    # Reset index to avoid KeyError
    X = X.reset_index(drop=True)
    y = y.reset_index(drop=True)

    skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state=random_state)
    scores = []

    for train_idx, val_idx in skf.split(X, y):
        X_train_cv, X_val_cv = X.iloc[train_idx], X.iloc[val_idx]
        y_train_cv, y_val_cv = y.iloc[train_idx], y.iloc[val_idx]

        model.fit(X_train_cv, y_train_cv)
        y_pred = model.predict(X_val_cv)
        scores.append(f1_score(y_val_cv, y_pred, average="macro"))

    return np.array(scores)

# Note: building wrappers for sklearn-like models
# expensive if running XGBoost/LightGBM multiple times; consider using smaller n_estimators during CV.
print("Running CV for XGBoost (this may take a while)...")
xgb_cv = cv_score_model(XGBClassifier(objective='multi:softprob', num_class=len(unique), n_estimators=100, random_state=random_state), X_train, y_train)
print("XGBoost CV macro-F1: mean {:.4f} ± {:.4f}".format(xgb_cv.mean(), xgb_cv.std()))

print("Running CV for LightGBM...")
lgb_cv = cv_score_model(lgb.LGBMClassifier(objective='multiclass', num_class=len(unique), n_estimators=150, random_state=random_state), X_train, y_train)
print("LightGBM CV macro-F1: mean {:.4f} ± {:.4f}".format(lgb_cv.mean(), lgb_cv.std()))

print("Running CV for KNN (fast)...")
knn_cv = cross_val_score(knn_pipeline, X_train, y_train, cv=skf, scoring='f1_macro', n_jobs=-1)
print("KNN CV macro-F1: mean {:.4f} ± {:.4f}".format(knn_cv.mean(), knn_cv.std()))

print("Running CV for SVM (may be slow)...")
svm_cv = cross_val_score(svm_pipeline, X_train, y_train, cv=skf, scoring='f1_macro', n_jobs=-1)
print("SVM CV macro-F1: mean {:.4f} ± {:.4f}".format(svm_cv.mean(), svm_cv.std()))

print("\nDone.")
```
